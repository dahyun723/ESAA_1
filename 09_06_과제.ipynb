##DataFrame 삭제
drop()메서드를 이용
DataFrame.drop(lavels=None, axis=0, index=None, colums=None, level=None, inplace= False, errors='raise')

1. axis 파라미터

axis 0 은 로우방향 축, axis 1은 칼럼 방향 축

axis=0, labes에 인덱스 입력: 해당 인덱스의 로우 드롭

axis=1, labels에 원하는 칼럼명 입력: 지정된 칼럼 드롭

```python
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

2. inplace 파라미터
디폴트값: False
inplace = False : 반환 자기 자신의 DataFrame 데이터 삭제 X, 삭제된 결과값
inplace = True : 자기 자신의 DataFrame 데이터 삭제 O, 반환값: None
* 유의: inplace = True인 상태로 반환 값을 다시 자신의 DataFrame 객체로 할당하면 안 됨

```python
titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, errors='ignore', inplace=True)
print('inplace=True로 drop 후의 데이터프레임:', titanic_df.head(3))
```

```python
# 데이터프레임의 열 이름 확인
print("현재 데이터프레임의 열 이름들:")
print(titanic_df.columns)

# 삭제할 열 이름 목록
columns_to_drop = ['Age_0', 'Age_by_10', 'Family_No']

# 데이터프레임에 존재하는 열만 필터링
existing_columns_to_drop = [col for col in columns_to_drop if col in titanic_df.columns]

# 열 삭제 (존재하지 않는 열은 무시됨)
titanic_df.drop(columns=existing_columns_to_drop, axis=1, errors='ignore', inplace=True)

# 열 삭제 후 데이터프레임의 상위 3개 행 출력
print("\n열 삭제 후의 데이터프레임 상위 3개 행:")
print(titanic_df.head(3))
```

##reset_index()
새로운 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스 'index'라는 새로운 칼럼명으로
추가
(주로 인덱스가 연속된 int 숫자형 데이터가 아닌 경우 사용)
* 유의: Series에 reset_index()를 적용하면 Series가 아닌 DataFrame이 반환된다.
reset_index()의 parameter 중 drop=True로 설정하면
기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제(drop)된다

```python
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
```

```python
print('### before reset_index ###')
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:', type(value_counts))
new_value_counts = value_counts.reset_index(inplace=False)
print('### After reset_index ###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:', type(new_value_counts))
```

##DataFrame iloc[] 연산자

iloc[]는 위치(location) 기반 인덱싱 - 행, 열, 위치를 0을 출발점으로 하는 세로축, 가로축 좌표 정수값 지정
loc[ ]는 위치(location)기반 인덱싱 방식이다
- 행과 열의 좌표 위치에 해당하는 값으로 정숫값 또는 정수형의 슬라이싱, 팬시 리스트 값을 입력


## loc[] 연산자


loc[]는 명칭(label) 기반 인덱싱 - 데이터 프레임 인덱스 값으로 행 위치, 칼럼 명칭으로 열 위치 지정
loc[인덱스값,칼럼명]에서 데이터의 인덱스 값을 확인하지 않고 정수값을 입력할 경우 오류 발생


```python
data = {'Name':['Chulmin', 'Eunkyung', 'Jiwoong', 'Soobeom'],
        'Year':[2011,2016,2015,2015],
        'Gender':['Male','Female', 'Male','Male']
        }
data_df = pd.DataFrame(data, index=['one','two','three','four'])
data_df
```

```python
# 아래 코드는 오류를 발생시킴
data_df.iloc['Name']
```

```python
#아래 코드는 오류를 발생시킴
data_df.iloc['one',0]
```

```python
print("\n 맨 마지막 칼럼 데이터 [:,-1] \n", data_df.iloc[:,-1])
print("\n 맨 마지막 칼럼을 제외한 모든 데이터 [:,:-1] \n:", data_df.iloc[:,:-1])
```
